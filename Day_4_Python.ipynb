```python
import pandas as pd
import numpy as np

data ={
    "Energy Source":["Solar","Wind","Hydropower","GeoThermal","Biomass","Nuclear"],
    "Energy Consumption (MwH)":[1200,np.nan,2900,np.nan,2500,3200],
    "Cost (Million $)":[200,400,np.nan,150,250,np.nan]
}

energy_df=pd.DataFrame(data)

print("Original Energy data with missing values:")

```

```python
energy_df
```

- 1.1. Remove Rows with Missing Values
- we can remove the row that contains missing values uing dropna()
image.png

```python
cleaned_df = energy_df.dropna()
```

```python
print("\nData After Removing Rows with Missing Values")
cleaned_df
```

```python
# how to handle nan values
'''
always make a graph(histogram) to check whether it is skewed or not,
and replace the nan values with median if skewness exist
and replace with mean if data is symmetric,
if the data is categorical, then replace Nan with mode
'''
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
mean_con = energy_df['Energy Consumption'].mean()
mean_con
con_mean = energy_df['Cost (Million $)'].mean()
```

```python
energy_df['Energy Consumption'].fillna(mean_con, inplace =True)
```

```python
energy_df
```

```python
energy_df['Cost (Million $)'].fillna(con_mean, inplace = True)
energy_df
```
